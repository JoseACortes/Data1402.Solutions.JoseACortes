```python
def CheckRate_Alpha(dataset, field, criteria, c):
    positive_count = 0
    negative_count = 0
    other = 0
    Seg_Set = dataset[['signal', field]]
    for n, s in enumerate(Seg_Set[field]):
        if Seg_Set['signal'][n] == 1:
            if criteria(s, c):
                positive_count += 1
            else:
                other += 1
        elif Seg_Set['signal'][n] == 0:
            if criteria(s, c):
                negative_count += 1
            else:
                other += 1
    total = positive_count+negative_count+other
    return positive_count/total, negative_count/total
```

```python
def makerange(a, b, step):
    rangelist = list()
    c = a
    while c < b:
        rangelist.append(c)
        c += step
    return rangelist
```

```python
def ChartTry_Alpha(dataset, field, criteria, checkrange):
    z = list()
    y = list()
    x = list()
    for c in checkrange:
        y.append(CheckRate_Alpha(dataset, field, criteria, c)[0])
        z.append(CheckRate_Alpha(dataset, field, criteria, c)[1])
        x.append(c)
    return x, y, z
```
